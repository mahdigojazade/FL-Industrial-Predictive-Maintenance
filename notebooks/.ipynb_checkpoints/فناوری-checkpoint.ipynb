```python
import pandas as pd

data = pd.read_csv("train_FD001.txt", sep=" ", header=None)
data = data.dropna(axis=1)

print(data.head())
```

```python
cols = ['unit_id','cycle'] + \
       [f'op_setting_{i}' for i in range(1,4)] + \
       [f'sensor_{i}' for i in range(1,22)]

data.columns = cols

print(data.head())
```

```python
max_cycle = data.groupby('unit_id')['cycle'].max().reset_index()

max_cycle.columns = ['unit_id', 'max_cycle']

data = data.merge(max_cycle, on='unit_id')

data['RUL'] = data['max_cycle'] - data['cycle']

print(data[['unit_id', 'cycle', 'max_cycle', 'RUL']].head())
```

```python
data['label'] = (data['RUL'] < 30).astype(int)

print(data[['RUL', 'label']].head(40))
```

```python
client_1 = data[data['unit_id'] <= 30]

client_2 = data[(data['unit_id'] > 30) & (data['unit_id'] <= 70)]

client_3 = data[data['unit_id'] > 70]

print(len(client_1))
print(len(client_2))
print(len(client_3))
```

```python
features = [col for col in data.columns if 'sensor' in col]

X1 = client_1[features]
y1 = client_1['label']

X2 = client_2[features]
y2 = client_2['label']

X3 = client_3[features]
y3 = client_3['label']

print(X1.head())
print(y1.head())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X1_scaled = scaler.fit_transform(X1)
X2_scaled = scaler.fit_transform(X2)
X3_scaled = scaler.fit_transform(X3)

print(X1_scaled[:5])
```

```python
import torch

X1_tensor = torch.tensor(X1_scaled, dtype=torch.float32)
y1_tensor = torch.tensor(y1.values, dtype=torch.float32)

X2_tensor = torch.tensor(X2_scaled, dtype=torch.float32)
y2_tensor = torch.tensor(y2.values, dtype=torch.float32)

X3_tensor = torch.tensor(X3_scaled, dtype=torch.float32)
y3_tensor = torch.tensor(y3.values, dtype=torch.float32)

print(X1_tensor.shape)
print(y1_tensor.shape)
```

```python
from torch.utils.data import TensorDataset, DataLoader

dataset1 = TensorDataset(X1_tensor, y1_tensor)
dataset2 = TensorDataset(X2_tensor, y2_tensor)
dataset3 = TensorDataset(X3_tensor, y3_tensor)

loader1 = DataLoader(dataset1, batch_size=32, shuffle=True)
loader2 = DataLoader(dataset2, batch_size=32, shuffle=True)
loader3 = DataLoader(dataset3, batch_size=32, shuffle=True)

print(len(loader1))
```

```python
import torch.nn as nn

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()

        self.model = nn.Sequential(
            nn.Linear(21, 64),
            nn.ReLU(),

            nn.Linear(64, 32),
            nn.ReLU(),

            nn.Linear(32, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.model(x)

model = Net()

print(model)
```

```python
criterion = nn.BCELoss()

optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

print("Ready")
```

```python
epochs = 5

for epoch in range(epochs):

    total_loss = 0

    for X_batch, y_batch in loader1:

        optimizer.zero_grad()

        outputs = model(X_batch).squeeze()

        loss = criterion(outputs, y_batch)

        loss.backward()

        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1}, Loss: {total_loss:.4f}")
```

```python
correct = 0
total = 0

with torch.no_grad():

    for X_batch, y_batch in loader1:

        outputs = model(X_batch).squeeze()

        predicted = (outputs > 0.5).float()

        correct += (predicted == y_batch).sum().item()

        total += y_batch.size(0)

accuracy = correct / total

print("Accuracy:", accuracy)
```

```python
model1 = Net()
model2 = Net()
model3 = Net()
```

```python
def train_model(model, loader, epochs=3):

    criterion = nn.BCELoss()

    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

    for epoch in range(epochs):

        total_loss = 0

        for X_batch, y_batch in loader:

            optimizer.zero_grad()

            outputs = model(X_batch).squeeze()

            loss = criterion(outputs, y_batch)

            loss.backward()

            optimizer.step()

            total_loss += loss.item()

        print(f"Loss: {total_loss:.4f}")
```

```python
print("Training Client 1")
train_model(model1, loader1)

print("Training Client 2")
train_model(model2, loader2)

print("Training Client 3")
train_model(model3, loader3)
```

```python
global_model = Net()

global_dict = global_model.state_dict()

for key in global_dict.keys():

    global_dict[key] = (
        model1.state_dict()[key] +
        model2.state_dict()[key] +
        model3.state_dict()[key]
    ) / 3

global_model.load_state_dict(global_dict)

print("FedAvg Done")
```

```python
correct = 0
total = 0

with torch.no_grad():

    for X_batch, y_batch in loader1:

        outputs = global_model(X_batch).squeeze()

        predicted = (outputs > 0.5).float()

        correct += (predicted == y_batch).sum().item()

        total += y_batch.size(0)

accuracy = correct / total

print("Global Model Accuracy:", accuracy)
```
